## Load data into Pandas from CSV

```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
df
```

## Read data in Pandas

#### Read column names

```python
df.columns
```

#### Read only 'Name' column. Return only first 10 rows

```python
df['Name'][0:10]
```

#### Read 'Name', 'Type 1' and 'Type 2' columns. Return first 5 rows

```python
df[['Name', 'Type 1', 'Type 2']][0:5]
```

#### Find info for Pokemon with index no36

```python
df.iloc[36]

# Pokemon with index no36 is the Poison/Ground type Pokemon: Nidoqueen
```

#### Find only the name for the Pokemon with index no78

```python
df.iloc[78, 1]

# using iloc I specified the row with index 78 and selected column number 1
```

#### Find only Water type Pokemon. Return first 10 rows

```python
df.loc[df['Type 1'] == 'Water'][0:5]
```

#### Sort by name (asc)

```python
df.sort_values('Name')
```

#### Sort by name (desc)

```python
df.sort_values('Name', ascending = False)
```

#### Find all 'Mega' Pokemon

```python
df.loc[df['Name'].str.contains('Mega')][0:5] #Displaying only top 5 to save space, delete [0:5] to see all rows returned
```

#### Find all Pokemon that ARE NOT 'Mega'

```python
df.loc[~df['Name'].str.contains('Mega')]
```

## Edit dataframe

#### Create 'Total' column

```python
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
df.head(5)
```

#### Drop 'Total' column

```python
df = df.drop(columns=['Total'])
df.head(5)
```

#### Create 'Total' column using a different method

```python
df['Total'] = df.iloc[:,4:10].sum(axis=1)
df.head(5)
```

## Aggregate statistics

#### Group by 'Type 1', return mean values and sort by 'Attack' desc

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

#### Count how many Pokemon of each type ('Type 1' only)

```python
df['Count'] = 1
df.groupby(['Type 1']).count()['Count']
```
